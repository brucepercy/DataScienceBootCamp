```python
from sklearn.impute import SimpleImputer
import numpy as np

# Mean strategy to impute missing values with mean
imputer_mean = SimpleImputer(missing_values=np.nan, strategy='mean')

# Most frequent strategy to impute missing values with most frequent value
imputer_most_frequent = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

# Custom constant strategy to impute missing values with a specified constant
constant_value = 0  # You can replace this with any desired constant value
imputer_custom_constant = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=constant_value)
```
