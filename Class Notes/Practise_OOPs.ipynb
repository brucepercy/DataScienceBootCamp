```python
class BankAccount:
  def __init__(self, name, account_type, available_balance, withdraw_amount, deposit_amount):
    self.name = name
    self.account_type = account_type
    self.available_balance = available_balance
    self.withdraw_amount = withdraw_amount
    self.deposit_amount = deposit_amount

  def deposit(self, deposit_amount):
    new_balance= self.available_balance + deposit_amount
    print("Deposited:", deposit_amount)
    print('New balance:', new_balance)

  def withdraw(self, withdraw_amount):
    if withdraw_amount > self.available_balance:
      print("Low Balance.")
    else:
      new_balance = self.available_balance - self.withdraw_amount
      print("amount withdrawn:", withdraw_amount)
      print("New Balance:",new_balance )


  def current_balance(self):
    print("Current Balance:" , self.available_balance)
```

```python
# Online Shopping Cart: Develop an Item class and a ShoppingCart class.
# The ShoppingCart class should allow adding items, removing items, calculating the total cost,
# and applying discounts.

class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def remove_item(self, item):
        self.items.remove(item)

    def calculate_total_cost(self):
        total_cost = 0
        for item in self.items:
            total_cost += item.price
        return total_cost

    def apply_discount(self, discount_percentage):
        total_cost = self.calculate_total_cost()
        discount_amount = (discount_percentage / 100) * total_cost
        return total_cost - discount_amount
```

```python
# Vehicle Rental System: Build classes for Vehicle, Car, and Bike.
# Implement a rental system where customers can rent vehicles and the system can track availability.

class Vehicle:
  def __init__(self, make, model, vehicle_type):
    self.make = make
    self.model = model
    self.vehicle_type = vehicle_type

class Car(Vehicle):
  def __init__(self, make, model):
    super().__init__(make, model, "car")

class Bike(Vehicle):
  def __init__(self, make, model):
    super().__init__(make, model, "bike")

class RentalSystem:
  def __init__(self):
    self.available_cars = []
    self.available_bikes = []

  def add_vehicle(self, vehicle):
    if isinstance(vehicle, Car):
        self.available_cars.append(vehicle)
    elif isinstance(vehicle, Bike):
        self.available_bikes.append(vehicle)

  def remove_vehicle(self, vehicle):
    if isinstance(vehicle, Car):
      self.available_cars.remove(vehicle)
    elif isinstance(vehicle, Bike):
      self.available_bikes.remove(vehicle)

  def rent_vehicle(self, vehicle_type):
    if vehicle_type == "car":
      if len(self.available_cars) > 0:
        return self.available_cars.pop(0)
      else:
        return None
    elif vehicle_type == "bike":
      if len(self.available_bikes) > 0:
        return self.available_bikes.pop(0)
      else:
        return None

  def get_available_vehicles(self):
    available_vehicles = []
    available_vehicles.extend(self.available_cars)
    available_vehicles.extend(self.available_bikes)
    return available_vehicles
```
