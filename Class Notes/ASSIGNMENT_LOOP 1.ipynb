```python
# 3) Remove duplicate elements from a list without converting to tuple
# Method 2 using for loop

my_list = [1,1,2,3,4,5,5,6,7,6,5,454,3,4,5,3,5,5,5566,6,6,7,6,5,55,56,6]
#print original list
print("Original List")
print(my_list)

# Creating a temp list to input the number from my_list
temp_list =[]
# Each time it is going to check that the number is there in temp_list
# and then if its not there it is going to append the list
for i in my_list:
  if i not in temp_list:
    temp_list.append(i)

# After all the iterations are over and temp_list is created change the name back to my_list
my_list  = temp_list
print()
print("Duplicate values removed list")
print(my_list)
```

```python
# 3) Write a program to find greatest common divisor (GCD)
#    or highest common factor (HCF) of given two numbers

# take the input of two integers to calculate the HCF
num1 = int(input("Enter an integer: "))
num2 = int(input("Enter an integer: "))

# define a function
def hcf(num1, num2):

#finding the samll int of the use given inputs
  if num1 > num2:
    small_int =num2
  else:
    small_int =num1
  for i in range(1,small_int +1):
    if(num1 % i == 0) and (num2 % i == 0):
      hcf = i
  return hcf

int_hcf = hcf(num1,num2)
print("The highest Common Factor of", num1,"," , num2 , "is", int_hcf)
```

```python
# 6) Write a program to print number names of digits
#    in a number like
#    SAMPLE INPUT 231 		SAMPLE OUTPUT TWO THREE ONE

int = input("Enter a number you want to converd to words: ")

def print_value(digit):

  if digit == '0':
      print("Zero ")

  elif digit == '1':
      print("One ")

  elif digit == '2':
      print("Two ")

  elif digit=='3':
      print("Three")

  elif digit == '4':
      print("Four ")

  elif digit == '5':
      print("Five ")

  elif digit == '6':
      print("Six ")

  elif digit == '7':
      print("Seven ")

  elif digit == '8':
      print("Eight")

  elif digit == '9':
      print("Nine ")

def print_name(int):
  i = 0
  length = len(int)

  while i < length:
    print_value(int[i])
    i += 1

print_name(int)
```

```python
# 7) Write a program to print Fibonacci series till n terms take n =50

def fib(n):
  a,b = 0,1 # first wo terms of the series
  count = 0 # initializing the count 0

  while count < n:
    print(a, end =' ')
    c = a +b
    a = b
    b = c
    count +=1

n = 50
fib(n)
```

```python
# 9) Write a program to check if a number is Armstrong or not
number = int(input("Enter a number to check if its an Armstrong number or not: "))
def is_armstrong(number):
  digits = [int(digit) for digit in str(number)]
  length = len(digits)
  digit_sum = sum(digit ** length for digit in digits)
  return digit_sum == number
if number:
  if is_armstrong(number):
    print(number, "Is an Armstrong number")
  else:
    print(number, "Is NOT an Armstrong number")
```

```python
# 18) Write a Python program to check if a triangle is equilateral,
# isosceles or scalene. Take 3 inputs for each side

print("Check your trianlge is a equilateral, isosceles, scalene triangle")
print()
side1 = int(input("Enter side 1: "))
side2 = int(input("Enter side 2: "))
side3 = int(input("Enter side 3: "))
print()

if (side1 + side2 > side3) and (side1 + side3 > side2) and (side2 + side3 > side1):
  if side1 == side2 == side3:
    print("Equilateral triangle.")

  elif side1 == side2 or side1 == side3 or side2 == side3:
    print("Isosceles triangle.")

  else:
    print("Scalene triangle.")
```
