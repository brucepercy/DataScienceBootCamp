```python
# 5) find mean salary of people with age above 50

# 6) find mean salary of people with age below 20

# 7) find which education category has highest wage

# 8) find education and age for highest and lowest wage
```

```python
# 5) find mean salary of people with age above 50
df_cleaned.Wage[df_cleaned['Age'] > 50].mean()
```

```python
# 6) find mean salary of people with age below 20
df_cleaned.Wage[df_cleaned['Age'] < 20].mean()
```

```python
# 7) find which education category has highest wage
# First we need to find the sum for each category
# Then find out the max out of it
# df_cleaned.groupby('Education')['Wage'].max()

dictionary_of_mean_wage = {}
for k in df_cleaned.Education.value_counts().keys():
  dictionary_of_mean_wage[k] = df_cleaned.Wage[df['Education']== k].mean()

print(dictionary_of_mean_wage)
education_stats = df_cleaned.groupby('Education')['Wage'].agg(['count', 'mean'])

highest_wage_category = education_stats['mean'].idxmax()
highest_wage_value = education_stats['mean'].max()

print("Education category with the highest wage:", highest_wage_category)
print("Highest wage value:", highest_wage_value)
```

```python
# 8) find education and age for highest and lowest wage
print(df_cleaned[df_cleaned['Wage'] == df_cleaned.Wage.min()])
print("====================================")
print(df_cleaned[df_cleaned['Wage'] == df_cleaned.Wage.max()])
```
