```python
for i in range(10):  # looping through numbers from 0 to 9 (10 iterations)
    for j in range(0, 10 - i):  # loop to print spaces before stars
        print(' ', end='')  # printing a space character without moving to the next line
    for k in range(0, i + 1):  # loop to print asterisks in each row
        print('*', end=' ')  # print an asterisk followed by a space without moving to the next line
    print('\r')  # move to the next line after printing spaces and asterisks in the row
```

```python
# Outer loop runs from 0 to 9 (10 iterations)
for i in range(10):

    # first inner loop prints spaces before stars (range: 0 to i-1)
    for j in range(0, i):
        print(' ', end=' ')  # printing spaces accordingly

    # second inner loop prints stars (range: 10 to i+1 in reverse)
    for k in range(10, i, -1):
        print('*', end=' ')  # printing star and the spaces

    print('\r')  # move to the next line after each iteration of the outer loop
```

```python
# Generating Hollow Pyramid Pattern Using Stars

row = int(input('Enter number of rows required: '))

for i in range(row):
# This for loop iterates from 0 to row - 1. The variable i represents the current row number.
    for j in range(row-i):
        print(' ', end='') # printing space required and staying in same line
# This loop is responsible for printing the leading spaces before the asterisks.
# It iterates from 0 to row - i - 1 in order to print the appropriate number of spaces for each row, where i is the current row number.
# The print(' ', end='') statement is used to print a space character without moving to a new line.
    for j in range(2*i+1):
        if j==0 or j==2*i or i==row-1:
            print('*',end='')
        else:
            print(' ', end='')
# This loop handles the printing of asterisks and spaces to form the hollow part of the pyramid.
# It iterates from 0 to 2 * i, which is the number of characters needed in the current row.
# The if statement inside the loop checks three conditions:
        #j == 0 prints an asterisk at the beginning of the row.
        #j == 2 * i prints an asterisk at the end of the row.
        #i == row - 1 prints asterisks in the last row to form the base of the pyramid.

#For all other values of j within this range, a space is printed to create the hollow effect

    print() # printing new line
```
