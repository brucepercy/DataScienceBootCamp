```python
class School:
  def __init__(self, school_fname, school_lname):
    self.schoolfirstname = school_fname
    self.schoollastname = school_lname

  def printname(self):
    print('Class School: ', self.schoolfirstname, self.schoollastname)
```

```python
class Faculty(School):
  def __init__(self, program_name, program_strt_date):
    self.programname = program_name
    self.program_start_date = program_strt_date

  def printname(self):
   print('Faculty information fromm class facutly (School) : ', self.programname, self.program_start_date )
```

```python
class Teacher(Faculty):
  def __init__(self, teacher_fname, teacher_lname):
    self.teacherfirstname = teacher_fname
    self.teacherlastname = teacher_lname

  def printname(self):
    print("Instructor Name from Class Teacher (Faculty): ", self.teacherfirstname, self.teacherlastname)
```

```python
obj1 = School("Durham", "College")
obj2 = Faculty("Computer Programming", "Aug 7th")
obj3 = Teacher('Anurag', 'Edlabadkar')
obj4 = Teacher( "Bruce", "Percy")
```

```python
obj1.printname()
obj2.printname()
obj3.printname()
obj4.printname()
```

```python
# SUPER()
class car:
  def __init__(self, name, model):
    self.name = name
    self.model = model
  def print_details(self):
    print('class car: ', self.name , self.model)
```

```python
class audi(car):
  def __init__(self,name,model):
    super().__init__(name, model)
```

```python
obj1 = audi('audi 777', 777)
```

```python
obj1.print_details()
```

```python
class audi(car):
  def __init__(self,name,model,make):
    super().__init__(name, model)
    self.make = 1978

  def print_details(self):
    print(self.name,self.model, self.make)
```

```python
obj1 = audi('audi', 777, 1978)
```

```python
obj1.print_details()
```

```python
class Vehicle:
  def __init__(self, make, model, year, drivetrain, cc):
    self.make = make
    self.model = model
    self.year = year
    self.drivetrain = drivetrain
    self.cc = cc

  def Vehicle_details(self):
    print(self.make, self.model, self.year, self.drivetrain, self.cc)
```

```python
obj1 = Vehicle("Toyota", "Supra", 2003, "RWD", 2000)
obj2 = Vehicle("Ford", "Mustang", 1969, "RWD", 6400)
```

```python
obj1.Vehicle_details()
obj2.Vehicle_details()
```

```python
class Supercar(Vehicle):
  def __init__(self, make, model, year, drivetrain, cc):
    super().__init__(self, make, model, year, drivetrain, cc)
```

```python
obj3 = Supercar("Rimac", "Nevera", "2023", "AWD", "E")
```

```python
obj3.Vehicles_details()
```
