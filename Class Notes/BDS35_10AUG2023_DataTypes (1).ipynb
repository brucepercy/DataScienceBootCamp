```python
lst1 = ['Apple', 'Banana', 'Cherries', 'Dates', 'Eggfruit', 'Fig', 'Grapes', 'Honeyfruit', 'Icefruit']
# Here the indexof apple is 1
print(lst1[:len(lst1):])
#POSITIVR INDEXING POSITIVE SLICING
print(lst1[::-1])
#NEGATIVE INDEXING NEGATIVE SLICING
print(lst1[-1::-1])
#POSITIVE INDEX NEGATIVE SLICING
print(lst1[9::-1])
#NEGATIVE INDEXING POSITIVE SLICING
print(lst1[-9::1])
```
