```python
list_of_prime_numbers = []  # Initialize an empty list to store prime numbers.

# Loop through numbers from 2 to 99 (below 100).
for k in range(2, 100):
    flag = True  # Initialize a flag to represent if the number is divisible.

    # Check divisibility of 'k' with numbers from 2 to 'k-1'.
    for s in range(2, k):
        if k % s == 0:
            flag = False  # If 'k' is divisible, set flag to False.
            break  # No need to continue checking further divisors, exit the loop.

    if flag:
        list_of_prime_numbers.append(k)  # If flag is still True, 'k' is prime.

# Loop through the list of prime numbers and print each prime number.
for prime in list_of_prime_numbers:
    print(prime, 'is a prime number')
```

```python
# Write a program to find greatest common divisor (GCD)
# or highest common factor (HCF) of given two numbers.
# Get user input for two numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Define a function to calculate gcd of two numbers
def gcd(a, b):
    # Find minimum of the two numbers
    result = min(a, b)
    # Use a loop to check for gcd
    # Decrease result by 1 at each iteration until gcd is found
    while result:
        if a % result == 0 and b % result == 0:
            break
        result -= 1
    # Return the gcd of the two numbers
    return result

# Call the gcd function, passing in the two user inputs
ans_gcd = gcd(num1, num2)

# Print the result of the gcd calculation to the user
print(ans_gcd)
```

```python
def count_const_vowels(words):
    vowels_list = ['a', 'e', 'i', 'o', 'u']
    count_vowels = 0
    count_const = 0
    for k in words:
        if k.lower() in vowels_list:  # Convert k to lowercase for case-insensitive matching
            count_vowels += 1
        else:
            count_const += 1
    print("Count of Consonants:", count_const)
    print("Count of Vowels:", count_vowels)

input_string = input("Enter a string: ")
count_const_vowels(input_string)
```

```python
# # Write a Python function to check whether a number is "Perfect" or not.
# # According to Wikipedia : In number theory, a perfect number is a positive
# integer that is equal to the sum of its proper positive divisors,
#  that is, the sum of its positive divisors excluding the number
#  itself (also known as its aliquot sum). Equivalently, a perfect number
#   is a number that is half the sum of all of its
#   positive divisors (including itself).
# # Example : The first perfect number is 6, because 1, 2, and 3 are
# its proper positive divisors, and 1 + 2 + 3 = 6. Equivalently,
# the number 6 is equal to half the sum of all its positive divisors:
#  ( 1 + 2 + 3 + 6 ) / 2 = 6.
#  The next perfect number is 28 = 1 + 2 + 4 + 7 + 14.
#   This is followed by the perfect numbers 496 and 8128.

n = int(input("Enter Number: "))

sum1 = 0
for i in range(1, n):
  if(n % i == 0 ):
    print(i)
    sum1 = sum1 + i
if (sum1 == n):
  print("The number is a perfect number")
else:
  print("The number is not a perfect number")

```

```python
# 4) Write a program to print twin primes less than 1000.
#  If two consecutive odd numbers are both
#  prime then they are known as twin primes
def is_prime(num):
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    return True

twin_primes = []

for num in range(2, 1000):
    if is_prime(num) and is_prime(num + 2):
        twin_primes.append((num, num + 2))

print("Twin Prime Numbers less than 1000:")
for prime_pair in twin_primes:
    print(prime_pair)
```

```python
# 2) Write a program to find out the prime factorization of a number.
# Example: prime factorization  of 56 2 × 2 × 2 × 7
# We will divide the number until it becomes 1

n=int(input("Enter the Number for prime factorization"))

def prime_factors(num):
  factors=[]
  divisor=2
  while num>1:
    while num%divisor==0:
      num=num/divisor
      factors.append(str(divisor))
    divisor=divisor+1
  return factors

def show_prime_factorization(n):
  factors=prime_factors(n)
  return "x".join(factors)

show_prime_factorization(83)
```
