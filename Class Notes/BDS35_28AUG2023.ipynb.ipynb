  A collection of modules that are linked together is also known as a Python library. It has code bundles that can be used again and again in different programs. For programmers, it makes Python programming easier and simpler. Since then, we will not need to compose the same code for various projects.

```python
from numpy.core.shape_base import vstack
v  = np.eye(5)
print(v)
```

```python
v = np.random.random((10,10,10))
print(v)
```

```python
v = np.random.random((10,10))
v_max, v_min = v.max(), v.min()
print(v_max, v_min)
```

```python
v = np.random.random(30)
v_mean = v.mean()
print(v_mean)
```

```python
v = np.ones((10,10))
#print(v)
v[1:-1, 1:-1] = 0 # first 1:-1 is row and second 1:-1 is column
print(v)
```

```python
v = np.zeros((10,10))
#print(v)
v[1:-1, 1:-1] = 1 # first 1:-1 is row and second 1:-1 is column

print(v)
```

```python
v = np.ones((5,5))
print(v)
v = np.pad(v, pad_width=1, mode='constant', constant_values='0')
print(v)
```

```python
# Create a 5x5 matrix with values 1,2,3,4

# *   List item
# *   List item

# just below the diagonal
# v = np.num((5,5))
# num = num +1

# Create a 5x5 matrix filled with zeros
matrix = np.zeros((5, 5))

# Fill the matrix diagonal with values 1, 2, 3, 4, and 5
values = [1, 2, 3, 4, 5]
np.fill_diagonal(matrix, values)

print(matrix)
```

```python
z = np.diag(1+np.arange(4), k =-1) # number of rows stays the same but colums change
print(z)
```
