```python
# List
# List is represented by square bracket []
# list is mutable


list1 = [12, 23, 45, 56, 65, 76, 34, 90, 0]
print(list1)
print(type(list1))
```

```python
# Tuple
# Tuples is represented by circular bracket ()
# Tuple is immutable

tuple1 = (1,2,3,4,5,6,7)
print(tuple1)
print(type(tuple1))
```

```python
# Set
# Set is a combination of unique values
# Set is collection of un ordered elements
# Set is represented by curly brackets
# set is mutable

set1 = {'a', 'b', 'c', 'D', 'F', '$','#', 1, 2, 3, '2' }
print(set1)
print(type(set1))
```
