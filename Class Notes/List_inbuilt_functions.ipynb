```python
a_list = [11,2,3,4,5,6,7]
b_list = a_list
c_list = a_list.copy()

print ("Before Operation Actual List", a_list)
print ("Before Operation Equal List", b_list)
print ("Before Operation Copy List", c_list)
print("=============================================")

# Make a operation on a_list
a_list.append(8)
print ("After Operation Actual List", a_list)
print ("After Operation Equal List", b_list)
print ("After Operation Copy List", c_list)
```

```python
a_list = [1,1,1,1,1,2,3,4,5,5,6,6,66,66,'a', 'b']
print(a_list.count(1))
print(a_list.count(66))
print(a_list.count('a'))
```

```python
# Extend function
a_list = [11,2,3,4,5,6,7]
a_list.append([8,9,10])
print("Extend function", a_list)
print("Extend function len", len(a_list))
```

```python
# Extend function
a_list = [11,2,3,4,5,6,7]
a_list.extend([8,9,10])
print("Extend function", a_list)
print("Extend function len", len(a_list))
```

```python
# Print the 9 in the list
a_list = [11,2,3,4,5,6,7]
a_list.append([8,9,10])

print(a_list)
print(a_list[7][1])
print("Extend function len", len(a_list))
```

```python
# Print the 9 in the list
a_list = [11,2,3,4,5,6,7]
a_list.append([8,9,10])

print(a_list)
print(a_list[7])
print(a_list[7][::-1])
print("Extend function len", len(a_list))
```
