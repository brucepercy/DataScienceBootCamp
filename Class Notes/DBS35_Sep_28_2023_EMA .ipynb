```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the CSV file into a pandas DataFrame
df = pd.read_csv('Stock_data.csv')

# Assuming your CSV contains a column named 'Close' for the closing prices

# Step 2: Ask the user for start and end dates
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Convert date columns to datetime type
df['Gmt time'] = pd.to_datetime(df['Gmt time'])
df.set_index('Gmt time', inplace=True)

# Filter data for the specified date range
df = df[start_date:end_date]

# Step 3: Calculate the 12-period and 26-period EMAs
df['12-day EMA'] = df['Close'].ewm(span=12, adjust=False).mean()
df['26-day EMA'] = df['Close'].ewm(span=26, adjust=False).mean()

# Step 4: Plot the Close price and EMAs in a line chart
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['12-day EMA'], label='12-day EMA')
plt.plot(df['26-day EMA'], label='26-day EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price and EMAs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Reading the dataset
df = pd.read_csv('Stock_data.csv')

# Convert 'Gmt time' column to datetime
df['Gmt time'] = pd.to_datetime(df['Gmt time'])

# Ask the user for start and end dates in the formatt
start_date = input("Enter the start date in this format (YYYY-MM-DD): ")
end_date = input("Enter the end date in this format (YYYY-MM-DD): ")

# Convert the start and end dates to datetime
start_date = datetime.strptime(start_date, '%Y-%m-%d')
end_date = datetime.strptime(end_date, '%Y-%m-%d')

# Set 'Gmt time' as the index for easy date-based filtering
df.set_index('Gmt time', inplace=True)
# need to learn
# When using inplace=True, the original DataFrame (df in this case) is modified directly, and the change is permanent.
# When using inplace=False (or not specifying inplace), a new modified DataFrame is returned, and the original DataFrame (df in this case) remains unchanged.

# Filter the DataFrame based on the start and end dates
df_filtered = df[(df.index >= start_date) & (df.index <= end_date)]

# Calculate the 12-period EMA and 26-period EMA for the Close price
df_filtered['EMA_12'] = df_filtered['Close'].ewm(span=12).mean()
df_filtered['EMA_26'] = df_filtered['Close'].ewm(span=26).mean()

plt.figure(figsize=(12, 6))
# plt.xlabel('Date')
# plt.ylabel('Price')
# plt.title('Close Price and EMAs')
plt.plot(df_filtered.index, df_filtered['Close'], label='Close')
plt.plot(df_filtered.index, df_filtered['EMA_12'], label='EMA_12')
plt.plot(df_filtered.index, df_filtered['EMA_26'], label='EMA_26')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Close Price and EMAs')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Reading the dataset
df = pd.read_csv('Stock_data.csv')

# Convert 'Gmt time' column to datetime
df['Gmt time'] = pd.to_datetime(df['Gmt time'])

# Ask the user for start and end dates in the format
start_date = input("Enter the start date in this format (DD-MM-YYYY): ")
end_date = input("Enter the end date in this format (DD-MM-YYYY): ")

# Convert the start and end dates to datetime
start_date = datetime.strptime(start_date, '%d-%m-%Y')
end_date = datetime.strptime(end_date, '%d-%m-%Y')

# Set 'Gmt time' as the index for easy date-based filtering
df.set_index('Gmt time', inplace=True)

# Filter the DataFrame based on the start and end dates
df_filtered = df[(df.index >= start_date) & (df.index <= end_date)]

# Calculate the 12-period EMA and 26-period EMA for the Close price
df_filtered['EMA_12'] = df_filtered['Close'].ewm(span=12).mean()
df_filtered['EMA_26'] = df_filtered['Close'].ewm(span=26).mean()

# Extract day, month, year
df_filtered['Day'] = df_filtered.index.day
df_filtered['Month'] = df_filtered.index.month
df_filtered['Year'] = df_filtered.index.year

# Plot with date in the format 'DD-MM-YYYY'
plt.figure(figsize=(12, 6))
plt.plot(df_filtered.index, df_filtered['Close'], label='Close')
plt.plot(df_filtered.index, df_filtered['EMA_12'], label='EMA_12')
plt.plot(df_filtered.index, df_filtered['EMA_26'], label='EMA_26')
plt.xlabel('Date (DD-MM-YYYY)')
plt.ylabel('Price')
plt.title('Close Price and EMAs')
plt.legend()
plt.show()
```
