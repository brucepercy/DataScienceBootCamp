```python
class university_student:
  def __init__(self, fname, mname, lname, gender, email, contact):
    self.fname = fname
    self.mname = mname
    self.lname = lname
    self.gender = gender
    self.email = email
    self.contact = contact
    print(self.fname, self.mname, self.lname, self.gender, self.email, self.contact)
```

```python

class Vehicle:
  def __init__(self, make, model, year, drivetrain, cc):
    self.v_make = make
    self.v_model = model
    self.v_year = year
    self.v_drivetrain = drivetrain
    self.v_cc = cc

  def Vehicle_details(self):
    print(self.v_make, self.v_model, self.v_year, self.v_drivetrain, self.v_cc)
```

```python
# Great Grand Parent
class transportation:
	pass

# Grand Parent
class air(transportation):
	pass
class water(transportation):
	pass
class land(transportation):
	pass

# Parent
class wheeler2(land):
	pass
class wheeler3(land):
	pass
class wheeler4(land):
	pass
class wheeler6(land):
	pass
class wheeler8(land):
	pass
class wheeler10(land):
	pass
class wheeler12(land):
	pass

# Child
class gas(wheeler4):
	pass
class petrol(wheeler4):
	pass
class diseal(wheeler4):
	pass
class electri(wheeler4):
	pass
class battery(wheeler4):
	pass
class solar(wheeler4):
	pass
```

```python
# Great Grand Parent
class Vehicle:
	pass

# Grand Parent
class Car(Vehicle):
	pass
class Bike(Vehicle):
	pass

# Parent
class Sedan(Car):
	pass
class Sports(Car):
	pass
class Luxury(Car):
	pass

# Child
class Supercar(Sports):
	pass
class Coupe(Sports):
	pass
class Cabriolet(Sports):
	pass
```
