```python
# Encapsulation

class transport:
  def __init__(self, brand, model, year):
    self.brandofvehicle = brand
    self.modelofvehicle = model
    self.yearofmfg = year

  def print_details(self):
    print(self.brandofvehicle, self.modelofvehicle, self.yearofmfg)
```

```python
obj1 = transport("Toyota", "Supra", "2001")
```

```python
obj1.print_details()
```

```python
# Encapsulation

class transport:
  def __init__(self, brand, model, year):
    self.brandofvehicle = brand
    self.modelofvehicle = model
    self.yearofmfg = year

class cars(transport):
  def __init__(self,brand, model, year,  wheeltrain, fuel):
    super().__init__(brand, model, year)
    self.wheeltrainofcar = wheeltrain
    self.fueltype = fuel

  def print_details(self):
    print(self.brandofvehicle, self.modelofvehicle, self.yearofmfg, self.wheeltrainofcar, self.fueltype)
```

```python
obj2 = cars("Toyota", "Supra", "2001", "RWD", "Gas")
```

```python
obj2.print_details()
```
