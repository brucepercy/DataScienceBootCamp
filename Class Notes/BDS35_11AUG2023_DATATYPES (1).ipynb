```python
# Nested If

a = 155

if (a < 200):
  print('if (a < 200):')
  if (a < 175):
    print('if (a < 175):')
    if (a < 150):
      print('if (a < 150):')
      if (a < 125):
        print('if (a < 125):')
        if (a < 100):
          print('if (a < 100):')
          if (a < 75):
            print('if (a < 75):')
          else:
            print('else if (a < 75):')
        else:
          print('else if (a < 100):')
      else:
        print('else if (a < 125):')
    else:
      print('else if (a < 150):')
  else:
    print('else if (a < 175):')
else:
  print('else if (a < 200):')

```
