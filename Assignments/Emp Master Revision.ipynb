```python
# Allow user to manage Employees
# Each employee must have the following details
# id,Name,salary,age,job poistion
# Add employee
# Delete by id
# Update name by id
# update salary by id
# update job position by id
# update age by id
# delete by name
# get_details by id
# get_details by name
# print all employees
# exit

# All these options should be available to the user
```

```python
class Employee:
    def __init__(self, emp_id, name, salary, age, position):
        self.emp_id = emp_id
        self.name = name
        self.position = position
        self.salary = salary
        self.age = age

    def display_info(self):
        print("Employee ID:", self.emp_id)
        print("Name:", self.name)
        print("Position:", self.position)
        print("Salary:", self.salary)
        print("Age:", self.age)


class EmployeeController:
    def __init__(self):
        self.all_employees_list = []

    def add_employee(self, employee):
        self.all_employees_list.append(employee)

    def delete_by_id(self, the_id_to_delete):
        removed = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_delete:
                self.all_employees_list.remove(employee)
                removed = True
                break
        if removed:
            print("Employee Removed Successfully")
        else:
            print("No Employee Found")

    def update_name_by_id(self, the_id_to_update, the_name_to_update):
        updated = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_update:
                employee.name = the_name_to_update
                updated = True
                break
        if updated:
            print("Employee name updated Successfully")
        else:
            print("No such employee found")

    def update_salary_by_id(self, the_id_to_update, the_salary_to_update):
        updated = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_update:
                employee.salary = the_salary_to_update
                updated = True
                break
        if updated:
            print("Employee Salary Updated Successfully")
        else:
            print("No Such Employee found")

    def get_details_by_id(self, the_id_for_details):
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_for_details:
                employee.display_info()

    def get_details_by_name(self, the_name_for_details):
        all_employees_with_same_name = []
        for employee in self.all_employees_list:
            if employee.name == the_name_for_details:
                all_employees_with_same_name.append(employee)

        if len(all_employees_with_same_name) == 0:
            print("No employee found in record")
        elif len(all_employees_with_same_name) == 1:
            all_employees_with_same_name[0].display_info()
        elif len(all_employees_with_same_name) > 1:
            print("Found multiple employees with the same name, Please enter employee id")
            for employee in all_employees_with_same_name:
                print(employee.emp_id)
            id_of_the_name_for_details = int(input("Please enter the id: "))
            for employee in all_employees_with_same_name:
                if employee.emp_id == id_of_the_name_for_details:
                    employee.display_info()

    def delete_by_name(self, the_name_to_remove):
        all_employees_with_same_name = []
        for employee in self.all_employees_list:
            if employee.name == the_name_to_remove:
                all_employees_with_same_name.append(employee)
        if len(all_employees_with_same_name) == 0:
            print("No Such Employee Found in Record")
        elif len(all_employees_with_same_name) == 1:
            self.all_employees_list.remove(all_employees_with_same_name[0])
            print("Record Removed Successfully")
        elif len(all_employees_with_same_name) > 1:
            print("Found Multiple employees with same name please choose an ID you would like to Remove")
            for employee in all_employees_with_same_name:
                print(employee.emp_id)
            id_of_the_name_for_details = int(input("Please enter the id: "))
            self.delete_by_id(id_of_the_name_for_details)

    def update_age_by_id(self, the_id_to_update, new_age):
        updated = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_update:
                employee.age = new_age
                updated = True
                break

        if updated:
            print("Employee Age Updated Successfully")
        else:
            print("No such employee found")

    def update_position_by_id(self, the_id_to_update, new_position):
        updated = False
        for employee in self.all_employees_list:
            if employee.emp_id == the_id_to_update:
                employee.position = new_position
                updated = True
                break
        if updated:
            print("Employee Position Updated Successfully")
        else:
            print("No such employee found")

    def print_all_employee_details(self):
        for employee in self.all_employees_list:
            employee.display_info()
            print("===========================")
```

```python
controller_class_object = EmployeeController()

while True:
    print("Please choose the option")
    print("1) Add Employee")
    print("2) Delete Employee by ID")
    print("3) Delete Employee by Name")
    print("4) Print All Employee details")
    print("5) Update Name By id")
    print("6) Update Salary By id")
    print("7) Update Age By id")
    print("8) Update Position By id")
    print("9) Print All details by id")
    print("10) Print All details by name")
    print("11) Exit")
    choice = input("Please choose an option: ")

    if choice == '1':
        name = input("Please enter Employee name: ")
        salary = input("Please enter Employee salary: ")
        age = input("Please enter Employee age: ")
        position = input("Please enter Employee position: ")
        id = len(controller_class_object.all_employees_list)
        employee_object = Employee(id, name, salary, age, position)
        controller_class_object.add_employee(employee_object)

    elif choice == "2":
        id_to_remove = int(input("Please enter Employee ID you want to remove: "))
        controller_class_object.delete_by_id(id_to_remove)

    elif choice == "3":
        name_to_remove = input("Please enter Employee Name you want to remove: ")
        controller_class_object.delete_by_name(name_to_remove)

    elif choice == "4":
        controller_class_object.print_all_employee_details()

    elif choice == "5":
        id_to_update = int(input("Please enter Employee Id you want to Update: "))
        name_to_update = input("Please enter Name you want to Update the Employee name to: ")
        controller_class_object.update_name_by_id(id_to_update, name_to_update)

    elif choice == "6":
        id_to_update = int(input("Please enter Employee Id you want to Update: "))
        salary_to_update = input("Please enter Salary you want to Update the Employee name to: ")
        controller_class_object.update_salary_by_id(id_to_update, salary_to_update)


    elif choice == "7":
      id_to_update = int(input("Please enter Employee Id you want to Update: "))
      age_to_update = int(input("Please enter Age you want to Update the Employee age to: "))
      controller_class_object.update_age_by_id(id_to_update, age_to_update)

    elif choice == "8":
      id_to_update = int(input("Please enter Employee Id you want to Update: "))
      position_to_update = input("Please enter Position you want to Update the Employee to: ")
      controller_class_object.update_position_by_id(id_to_update, position_to_update)

    elif choice == "9":
        id_to_print_details = int(input("Please enter Employee Id you want to details of: "))
        controller_class_object.get_details_by_id(id_to_print_details)

    elif choice == "10":
        name_to_print_details = input("Please enter Employee name you want to details of: ")
        controller_class_object.get_details_by_name(name_to_print_details)

    elif choice == "11":
        print("Bye!")
        break
```
