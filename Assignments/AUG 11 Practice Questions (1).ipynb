```python
# Question : Green; car is allowed to go, Yellow; Car has to wait, Red; Car has to stop, other; Unrecognized
input_message = input("Please enter the color: ").lower()
if input_message == 'green':
    print('Car is allowed to go')
elif input_message == 'yellow':
    print('Car has to wait')
elif input_message == 'red':
    print('Car has to stop')
else:
    print('Unrecognized Symbol !')
```
