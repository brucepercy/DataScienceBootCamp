```python
"""
1.A company decided to give bonus of 5% to employee if his/her year of service is more than 5 years.
Ask user for their salary and year of service and print the net bonus amount.
"""
salary = float(input("Please enter your salary: "))
years = float(input("Years worked in the company: "))
print()
print("The salary you have entered is", salary)
print("Years you have worked is", years)
bonus_salary = salary * 1.05
if years >= 5:
  print("Your salary including 5% bonus is", bonus_salary)
else:
  print("Your salary is", salary)
```

```python
'''
2.Take values of length and breadth of a rectangle from user and check
 if it is square or not.
'''
breadth = float(input("Enter the breadth: "))
length = float(input("Enter the length: "))

if breadth == length:
    print("This is a square")
else:
    print("This is a rectangle")
```

```python
"""
4.A shop will give discount of 10% if the cost of purchased quantity is
more than 1000.
Ask user for quantity
Suppose, one unit will cost 100.
Judge and print total cost for user.
"""


quantity = int(input("Enter the quantity: "))
unit_price = 100

if (quantity * unit_price> 1000):
  print("Your Total cost inlcluding 10% off is :",quantity *  unit_price * 0.9)
else:
  print("Your total cost is: ", quantity *  unit_price )
```

```python
"""
5.A school has following rules for grading system:
a. Below 25 - F b. 25 to 45 - E c. 45 to 50 - D d. 50 to 60 - C e. 60 to 80 - B f.
Above 80 - A Ask user to enter marks and print the corresponding grade.
"""

mark = float(input("Please enter your marks: "))
if mark == mark >100 or mark < 0:
    print('Error: Mark should be between 0 and 100 only')
elif mark == mark < 25:
    print('Your Grade : F')
elif mark == mark >= 25 and mark < 45:
    print('Your Grade : E')
elif mark == mark >= 45 and mark < 50:
    print('Your Grade : D')
elif mark == mark >= 50 and mark < 60:
    print('Your Grade : C')
elif mark == mark >= 60 and mark < 80:
    print('Your Grade : B')
elif mark == mark >= 80 and mark <= 100:
    print('Your Grade : A')
```

```python
"""
6.Take input of age of 3 people by user and determine oldest and youngest among them.
"""

age1 = int(input("Please enter an age: "))
age2 = int(input("Please enter an age: "))
age3 = int(input("Please enter an age: "))

tuple1 = (age1,age2, age3)
print()
sorted_tuple = tuple(sorted(tuple1))
print(sorted_tuple)

youngest = sorted_tuple[0]
oldest = sorted_tuple[2]

print()
print("Youngest among them is", youngest, "years old")
print("Oldest among them is", oldest, "years old")
```

```python
"""8.A student will not be allowed to sit in exam if his/her attendence is
 less than 75%. Take following input from user Number of classes held
  Number of classes attended. And print percentage of class attended Is
  student is allowed to sit in exam or not."""


cls_held = int(input("Enter number of class held: "))
cls_att = int(input("Enter number of class attended: "))

percentage_cls_att = cls_att/cls_held * 100
print(percentage_cls_att)

if percentage_cls_att >= 75:
  print("The student is allowed to sit in exam.")
else:
  print("The student is not allowed to sit in exam.")
```

```python
"""
9.Modify the above question to allow student to sit if he/she has medical
 cause. Ask user if he/she has medical cause
 or not ( 'Y' or 'N' ) and print accordingly.
 """

cls_held = int(input("Enter number of classes held: "))
cls_att = int(input("Enter number of classes attended: "))

percentage_cls_att = cls_att / cls_held * 100
print("The percentage of your attendance is",round(percentage_cls_att))

if percentage_cls_att >= 75:
    print("The student is allowed to sit in the exam.")
else:
    medical = input("Absence due to medical cause? \n"
                   "Enter 'y' for yes and 'n' for no \n ").lower()
    if medical == 'y':
        print("The student is allowed to sit in the exam.")
    else:
        print("The student is not allowed to sit in the exam.")
```
